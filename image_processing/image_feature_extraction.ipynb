## Comenzamos el uso de librerias, comenzamos con Open CV

```python
#Cargamos la imagen 
imagen = cv2.imread('gato.webp')
```

```python
#Convertimos la imagen a RGB
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Creamos los subplots
fig, axs = plt.subplots(1, 2, figsize=(7, 4))

# Plot de la imagen normal
axs[0].imshow(imagen)
axs[0].set_title('Original Image')

# Plot de la imagen RG
axs[1].imshow(imagen_rgb)
axs[1].set_title('RGB Image')

# Remove ticks from the subplots
for ax in axs:
	ax.set_xticks([])
	ax.set_yticks([])

# Display the subplots
plt.tight_layout()
plt.show()
```

```python
# Definimos el ángulo de rotación (en grados)
angulo_rotacion = 45
# Obtenemos el centro de la imagen
centro_imagen = (imagen.shape[1] // 2, imagen.shape[0] // 2)
# Creamos la matriz de rotación
matriz_rotacion = cv2.getRotationMatrix2D(centro_imagen, angulo_rotacion, 1.0)
# Rotamos la imagen
imagen_rotada = cv2.warpAffine(imagen_rgb, matriz_rotacion, (imagen.shape[1], imagen.shape[0]))

# Creamos la figura y el subplot
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
# Mostramos la primera imagen en el subplot 1
axs[0].imshow(imagen)
axs[0].set_title('Imagen Real')

# Mostramos la segunda imagen en el subplot 2
axs[1].imshow(imagen_rotada)
axs[1].set_title('Imagen Rotada')

for ax in axs:
	ax.set_xticks([])
	ax.set_yticks([])

# Ajustamos el layout
plt.tight_layout()

# Mostramos la figura
plt.show()
```

```python
# Aplicamos un filtro de desenfoque gaussiano a la imagen
imagen_desenfocada = cv2.GaussianBlur(imagen, (7, 7), 0)

# Creamos la figura y el subplot
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(imagen)
axs[0].set_title('Imagen original')

axs[1].imshow(imagen_desenfocada)
axs[1].set_title('Imagen con desenfoque')

for ax in axs:
	ax.set_xticks([])
	ax.set_yticks([])

# Ajustamos el layout
plt.tight_layout()

# Mostramos la figura
plt.show()
```

```python
# Use the cvtColor() function to grayscale the image 
gray_image = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY) 

# converting to its binary form 
ret,thresh = cv2.threshold(gray_image,70,255,0)

# Creamos la figura y el subplot
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

axs[0].imshow(imagen)
axs[0].set_title('Imagen original')

axs[1].imshow(gray_image, cmap='gray')
axs[1].set_title('Imagen con escala de grises')

axs[2].imshow(thresh, cmap='gray')
axs[2].set_title('Imagen binaria')

for ax in axs:
	ax.set_xticks([])
	ax.set_yticks([])

# Ajustamos el layout
plt.tight_layout()

# Mostramos la figura
plt.show()
```

```python
# Convertimos la imagen a escala de grises
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Aplicamos el detector de bordes Canny
bordes = cv2.Canny(imagen_gris, 100, 200)

# Creamos la figura y el subplot
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
# Mostramos la primera imagen en el subplot 1
axs[0].imshow(imagen)
axs[0].set_title('Imagen Real')

# Mostramos la segunda imagen en el subplot 2
axs[1].imshow(bordes)
axs[1].set_title('Imagen con bordes')

for ax in axs:
	ax.set_xticks([])
	ax.set_yticks([])

# Ajustamos el layout
plt.tight_layout()

# Mostramos la figura
plt.show()
```

```python
# Cargamos la imagen
imagen = Image.open('gato.jpg')

# Cambiamos el tamaño de la imagen
imagen_redimensionada = imagen.resize((200, 200))

# Rotamos la imagen
imagen_rotada = imagen.rotate(45)

# Creamos la figura y el subplot
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

axs[0].imshow(imagen)
axs[0].set_title('Imagen original')

axs[1].imshow(imagen_redimensionada)
axs[1].set_title('Imagen con redimensión')

axs[2].imshow(imagen_rotada)
axs[2].set_title('Imagen rotada')

for ax in axs:
	ax.set_xticks([])
	ax.set_yticks([])

# Ajustamos el layout
plt.tight_layout()

# Mostramos la figura
plt.show()
```

```python
# Creamos una fuente
fuente = ImageFont.truetype('arial.ttf', 150)

# Creamos un objeto para dibujar en la imagen
dibujo = ImageDraw.Draw(imagen)

# Añadimos el texto
dibujo.text((10, 10), "Soy gato Karen", font=fuente, fill=(255, 0, 0))

# Cargamos la marca de agua
marca_agua = Image.open('descarga.png')

# Posicionamos la marca de agua
posicion_marca_agua = (imagen.width - marca_agua.width, imagen.height - marca_agua.height)

# Pegamos la marca de agua en la imagen
imagen.paste(marca_agua, posicion_marca_agua)

# Guardamos la imagen con el texto y la marca de agua
imagen.save('imagen_con_texto_y_marca_agua.jpg')
```

```python
# Cargamos la imagen
imagen = Image.open('gato.jpg')

# Convertimos la imagen a escala de grises
imagen_gris = imagen.convert('L')

# Ajustamos el brillo de la imagen
imagen_brillo = imagen.point(lambda x: x * 1.2)

# Extraemos el canal rojo
canal_rojo = imagen.split()[0]

#Extraemos el canal verde
canal_verde = imagen.split()[1]

#Extraemos el canal azul
canal_azul = imagen.split()[2]


# Creamos la figura y el subplot
fig, axs = plt.subplots(2, 3, figsize=(10, 5))

axs[0][0].imshow(imagen)
axs[0][0].set_title('Imagen original')

axs[1][0].imshow(imagen_gris)
axs[1][0].set_title('Imagen a escala de grises')

axs[0][1].imshow(imagen_brillo)
axs[0][1].set_title('Imagen con brillo')

axs[1][1].imshow(canal_rojo)
axs[1][1].set_title('Canal rojo')

axs[0][2].imshow(canal_verde)
axs[0][2].set_title('Canal verde')

axs[1][2].imshow(canal_azul)
axs[1][2].set_title('Canal azul')

for ax in axs.flat:  # Itera a través de todos los subplots
    ax.set_xticks([])
    ax.set_yticks([])
    
# Ajustamos el layout
plt.tight_layout()

# Mostramos la figura
plt.show()
```

```python
# Cargamos la imagen y la convertimos a un arreglo
imagen = Image.open('gato.jpg')
arreglo_imagen = np.array(imagen)

# Definimos el tamaño del área central a borrar
# Por ejemplo, vamos a borrar un área que sea el 50% del ancho y alto de la imagen
porcentaje_borrado = 0.5

# Calculamos las dimensiones del área a borrar
alto, ancho = arreglo_imagen.shape[:2]
ancho_borrado = int(ancho * porcentaje_borrado)
alto_borrado = int(alto * porcentaje_borrado)

# Calculamos el punto de inicio del área a borrar
inicio_x = (ancho - ancho_borrado) // 2
inicio_y = (alto - alto_borrado) // 2

# Establecemos los píxeles en el área definida a blanco (o al color deseado)
# Para imágenes RGB
arreglo_imagen[inicio_y:inicio_y+alto_borrado, inicio_x:inicio_x+ancho_borrado] = [255, 255, 255]  # Blanco

# Convertimos el arreglo modificado de vuelta a una imagen
imagen_modificada = Image.fromarray(arreglo_imagen)

plt.imshow(imagen_modificada)
plt.title('Imagen sin centro')
plt.axis('off');
```
